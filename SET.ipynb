```
"""Plays a game of SET and calculates some probabilities with
repetitive simulation.
"""

import random

numbers=[1,2,3]
colors=['green','red','purple']
shading=['empty','striped','solid']
shapes=['diamond','oval','squiggle']

def mod(num1,num2):
  """Returns the number that when added 
  to 2 arguments will give 0 (mod 3).
  """
  return (3 - (num1+num2))%3


def create_deck():
  """Creates a new deck of SET cards."""
  deck=[]
  for number in range(3):
    for color in range(3):
      for fill in range(3):
        for shape in range(3):
          card=[number,color,fill,shape]
          deck.append(card)
  return deck
        
def get_third_card(c1, c2):
  """Returns the 3rd card that will complete the SET."""
  c3=[]
  for attribute in range(4):
    c3.append(mod(c1[attribute],c2[attribute]))
  return c3

def print_card(card):
  """Prints the 4 attributes of the card."""
  number=numbers[card[0]]
  color=colors[card[1]]
  shade=shading[card[2]]
  shape=shapes[card[3]]
  return_string=""
  return_string += str(number)+" "+color+" "+shade+" "+shape
  if number !=1:
    return_string += 's'
  print(return_string)
  
def add_cards(table,deck,num,monitor=True):
  """Adds cards to table from the deck."""
  for deal in range(num):
    card=random.choice(deck)
    table.append(card)
    deck.remove(card)
  if monitor:
    print("\nDealing ",num," cards...")
    print(len(table)," cards on the table.")
    print(len(deck)," cards in the deck.", )
    print("\nCards on Table:")
    for card in table:
      print_card(card)
  return(table,deck)
  
def set_to_discard(c1,c2,c3,table,discard):
  """Transfers cards from table to discard pile."""
  discard.append(c1)
  discard.append(c2)
  discard.append(c3)
  table.remove(c1)
  table.remove(c2)
  table.remove(c3)
    
def find_set(table,discard,monitor=True,discard_SETs=True):
  """Looks for SETS on the table, removes them,
  and counts how many were found.
  """
  if monitor:
    print("\nLooking for SETs...") 
  num_sets=0
  i=0
  j=1
  maxi=len(table)-2
  maxj=len(table)-1
  while i<=maxi:    #comparing the ith and jth cards on table
    while j<=maxj:
      c1=table[i]
      c2=table[j]
      c3=get_third_card(c1,c2)     
      if c3 in table:
        if monitor:
          print("SET!")
          print_card(c1)
          print_card(c2)
          print_card(c3)
        num_sets+=1
        if discard_SETs:
          set_to_discard(c1,c2,c3,table,discard)
          maxi-=3
          maxj-=3
      j+=1
    i+=1
    j=i+1
  if monitor:
    print("\nNumber of sets found:  ", num_sets)
  return num_sets
          
  
#table =  [[2, 0, 1, 0], [2, 2, 0, 2], [0, 0, 0, 2], [2, 2, 2, 1], [1, 0, 0, 1], [1, 1, 0, 2], [0, 1, 2, 1], [2, 1, 2, 2], [0, 1, 0, 2], [1, 2, 0, 1], [0, 2, 1, 0], [2, 0, 2, 0]]

#deck=[e for e in deck if e not in table]


def prob_some_SETS(num_cards,num_SETS,num_trials):
  """Calculates the probability there
  are num_SETS sets in num_cards randomly chosen
  cards.
  """
  SET_count=0
  for count in range(num_trials):
    table=[]
    deck=create_deck()
    discard=[]
    add_cards(table,deck,num_cards,False)
    if find_set(table,discard,False)==num_SETS:
      SET_count+=1
  return round((SET_count/num_trials)*100,3)  #return probability to 3 decimals
    
  
def Expected_Value(num_cards,num_trials):
  """   WIP
  Calculates average number of sets
  found in num_cards randomly chosen
  cards.
  """
  EV_sum=0
  for count in range(num_trials):
    EV=0
    table=[]
    deck=create_deck()
    discard=[]
    add_cards(table,deck,12,False)
    EV+=find_set(table,discard,False,False)
    EV_sum+=EV
  return EV_sum/num_trials

def play_SET():
  """Plays a game of SET and shows that
  remaining cards on table sum to zero mod 3.
  """
  print("\nLet's play SET!")
  table=[]  
  deck=create_deck()
  discard=[]
  new_cards=12
  while deck:
    table,deck=add_cards(table,deck,new_cards)
    num_sets=find_set(table,discard)
    if num_sets==0 and len(deck)>3:   #Need to add 3 more cards if you got them
      new_cards = 3
    elif len(deck)>12-len(table) and len(table)<12:  #add more cards if fewer than 12
      new_cards = 12-len(table)
    elif len(deck)>12-len(table) and len(table)==12:  #add more cards if you have 12 (and couldn't find a set)
      new_cards=3
    else:
      new_cards = len(deck)  #otherwise just add the remaining cards in the deck

  print("\n===GAME OVER===")
  print("Remaining cards on table:")
  for card in table:
    print_card(card)

  mod_sum=[0,0,0,0]
  for attribute in range(4):
    for card in table:
      mod_sum[attribute]+=card[attribute]
      mod_sum[attribute]=mod_sum[attribute]%3
  print("The mod 3 sum of each attribute of remaing cards should be 0.")
  print(mod_sum)

#print("EV:  ",Expected_Value(12,10000))  

print("Question 1")
print("Probability of 1 set out of 3 cards: ",prob_some_SETS(3,1,50000),"%")
print("\nQuestion 2")
print("Probability of no sets out of 12 cards: ",prob_some_SETS(12,0,50000),"%")
print("\nQuestion 3")
print("Probability of no sets out of 4 cards: ",prob_some_SETS(4,0,50000),"%")
print("\nQuestion 4")
print("Probability of 0 sets out of 5 cards: ",prob_some_SETS(5,0,50000),"%")
print("\nCase 1")
print("Probability of 2 sets out of 5 cards: ",prob_some_SETS(5,2,50000),"%")
print("\nCase 2")
print("Probability of 1 set out of 5 cards: ",prob_some_SETS(5,1,50000),"%")




play_SET()

```
